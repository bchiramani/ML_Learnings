```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading Data

```python
path_spotify='../content/spotify.csv'
data_spotify = pd.read_csv(path_spotify , index_col="Date", parse_dates=True) 
```

```python
path_flight_delay="../content/flight_delays.csv"
data_flight_delay=pd.read_csv(path_flight_delay,index_col='Month')
```

```python
path_insurance="../content/insurance.csv"
data_insurance=pd.read_csv(path_insurance)
```

### Data Visualization

```python
plt.figure(figsize=(10,5))
plt.title("Daily streaming on popular songs")
```

```python
plt.figure(figsize=(20,6))
plt.title("Daily streaming on popular songs")
sns.lineplot(data=data_spotify)
```

```python
plt.figure(figsize=(20,6))
plt.title("Daily streaming of Shape of You ")
sns.lineplot(data=data_spotify['Shape of You'])
```

```python
plt.figure(figsize=(20,6))
plt.title("Daily streaming of Shape of You and Despacito")

sns.lineplot(data=data_spotify['Shape of You'], label="Shape of You")
sns.lineplot(data=data_spotify['Despacito'], label="Despacito")

plt.xlabel("Date")
```

# Barplot and heatmaps

## Data Exploration

```python
data_flight_delay.head()
```

## Data Visualization

```python
plt.figure(figsize=(20,6))
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")
sns.barplot(x=data_flight_delay.index,y=data_flight_delay['NK'])
```

```python
plt.figure(figsize=(20,6))
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")
sns.barplot(x=data_flight_delay.index,y=data_flight_delay['NK'])
plt.ylabel("Arrival delay (in minutes)")
```

```python
plt.figure(figsize=(20,6))
plt.title("Average Arrival Delay for Each Airline, by Month")
sns.heatmap(data=data_flight_delay,annot=False)
```

```python
plt.figure(figsize=(20,6))
plt.title("Average Arrival Delay for Each Airline, by Month")
sns.heatmap(data=data_flight_delay,annot=True)
plt.xlabel("Airline")
```

# ScatterPlot

The scatterplot can suggests that there is positive correlation (when one increases the other increases too) or a negative correlation.
We can add a regression line , or any other line that best fit the data 

```python
plt.figure(figsize=(20,6))
sns.scatterplot(x=data_insurance['bmi'],y=data_insurance['charges'])
```

```python
plt.figure(figsize=(20,6))
sns.scatterplot(x=data_insurance['bmi'],y=data_insurance['charges'])
sns.regplot(x=data_insurance['bmi'],y=data_insurance['charges'])
```

```python
plt.figure(figsize=(20,6))
sns.scatterplot(x='bmi',y='charges', hue='smoker',data=data_insurance)
#sns.scatterplot(x=data_insurance['bmi'],y=data_insurance['charges'], hue=data_insurance['smoker'])
```

```python
sns.lmplot(x="bmi",y="charges",hue="smoker",data=data_insurance)
```

```python
sns.swarmplot(x=data_insurance['smoker'],y=data_insurance['charges'])
```
